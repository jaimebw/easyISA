```python
import numpy as np
import matplotlib.pyplot as plt 
from easyISA.equations import presure,temperature,density
```

```python
altitude = np.linspace(0,15000,1000)
```

```python
fig, ax = plt.subplots()
pressure = np.vectorize(presure.meters)
p = pressure(altitude)/1000 # kPa
ax.plot(p,altitude)
ax.set_ylabel("Altitude[m]")
ax.set_xlabel("Pressure[kPa]")
ax.invert_xaxis()
ax.grid()
```

```python
fig, ax = plt.subplots()
temperature_vec = np.vectorize(temperature.meters)
t = temperature_vec(altitude)# K
ax.plot(t,altitude)
ax.set_ylabel("Altitude[m]")
ax.set_xlabel("Temperature[K]")
ax.invert_xaxis()
ax.grid()
```

```python
fig, ax = plt.subplots()
density_vec = np.vectorize(density.meters)
rho = density_vec(altitude)# K
ax.plot(rho,altitude)
ax.set_ylabel("Altitude[m]")
ax.set_xlabel("Density[kg/m^3]")
ax.invert_xaxis()
ax.grid()
```
